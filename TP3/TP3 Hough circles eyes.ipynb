```python
#importamos las librer√≠as.
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# Leemos la imagen
#=================
img = cv.imread('eyes.jpg')
img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
#img = cv.imread('eyes.jpg')
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
g_gray = cv.GaussianBlur(gray, ksize=(5,5),sigmaX=1 )
```

```python
plt.imshow(g_gray, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Dibujo los c√≠rculos en los iris por la transformada de Hough
#===================================================================
img_out = img_rgb.copy()
circles = cv.HoughCircles(g_gray,cv.HOUGH_GRADIENT,1,minDist=90,param1=150,param2=16,minRadius=27,maxRadius=32)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # Dibujamos el c√≠rculo externo
    cv.circle(img_out,(i[0],i[1]),i[2],(0,255,0),2)
    # Dibujamos el centro del c√≠rculo
    cv.circle(img_out,(i[0],i[1]),2,(0,0,255),3)
cv.imshow('detected circles',img_out)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img_out, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Dibujo los c√≠rculos en las pupilas por la transformada de Hough
#===================================================================
img_out = img_rgb.copy()
circles = cv.HoughCircles(g_gray,cv.HOUGH_GRADIENT,1,minDist=85,param1=120,param2=15,minRadius=5,maxRadius=16)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # Dibujamos el c√≠rculo externo
    cv.circle(img_out,(i[0],i[1]),i[2],(0,255,0),2)
    # Dibujamos el centro del c√≠rculo
    cv.circle(img_out,(i[0],i[1]),2,(0,0,255),3)
cv.imshow('detected circles',img_out)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img_out, cmap='gray', vmin=0, vmax=255)
plt.show()
```
